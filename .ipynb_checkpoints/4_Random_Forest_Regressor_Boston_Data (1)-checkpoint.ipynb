# Carga de datos

```python
data_url = "/content/drive/MyDrive/Formación/Bootcamp - Geekshub - Machine Learning/Boston.csv"

raw_df = pd.read_csv(data_url, sep=",")
target = raw_df['medv']
data = raw_df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis',
       'rad', 'tax', 'ptratio', 'black', 'lstat']]
feature_names = ('crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis',
       'rad', 'tax', 'ptratio', 'black', 'lstat')

X_boston,y_boston = data, target # Create X matrix and y vector from the dataset.

df_boston = raw_df
```

# Modelos

# Random Forest Regressor

*   Training dataset (70%) --> 404 rows, 13 variables
*   Test dataset (30%) --> 102 rows, 13 variables

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Split the 'features' and 'income' data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_boston, 
                                                    y_boston, 
                                                    test_size = 0.2, 
                                                    random_state = 4)

print('Train and test sizes {} {}'.format(X_train.shape, X_test.shape))
```

## Train the model

```python
# Import Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest Regressor
reg = RandomForestRegressor()

# Train the model using the training sets 
reg.fit(X_train, y_train)
```

## Model Evaluation - Train data

```python
from sklearn import metrics
```

```python
data = pd.read_csv(data_url, sep=";")

# Extraer la columna 'Close' como la variable objetivo (y)
y = data['Close']

# Crear una instancia del modelo de Random Forest Regressor
reg = RandomForestRegressor()

# Entrenar el modelo utilizando los datos de la columna 'Close'
reg.fit(y.values.reshape(-1, 1), y)

# Predecir los valores de la variable objetivo utilizando los datos de entrenamiento
y_pred = reg.predict(y.values.reshape(-1, 1))

# Calcular las métricas de evaluación
r2 = metrics.r2_score(y, y_pred)
adj_r2 = 1 - (1 - r2) * (len(y) - 1) / (len(y) - 1 - 1)
mae = metrics.mean_absolute_error(y, y_pred)
mse = metrics.mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

# Imprimir las métricas
print('R^2:', r2)
print('Adjusted R^2:', adj_r2)
print('MAE:', mae)
print('MSE:', mse)
print('RMSE:', rmse)
```

```python
plt.hist(y_train)
```

```python
# Visualizing the differences between actual prices and predicted values
# Si tenemos una tendencia en la gráfica es que estamos cometiendo sesgo. (bias)
plt.scatter(y_train, y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Prices vs Predicted prices")
plt.show()
```

```python
# Checking residuals
# Si sigue una línea recta, es que no estamos cometiendo sesgo porque igual que nos equivocamos 'por arriba', nos equivocamos por 'abajo'
plt.scatter(y_pred,y_train-y_pred)
plt.title("Predicted vs residuals")
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.show()
```

## Model Evaluation - Test data

```python
# Predicting Test data with the model
y_test_pred = reg.predict(X_test)
```

```python
# Model Evaluation
acc_rf = metrics.r2_score(y_test, y_test_pred)
print('R^2:', acc_rf)
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_test, y_test_pred))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
print('MAE:',metrics.mean_absolute_error(y_test, y_test_pred))
print('MSE:',metrics.mean_squared_error(y_test, y_test_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

max_depth (int) – Maximum tree depth for base learners.

learning_rate (float) – Boosting learning rate (xgb’s “eta”)

n_estimators (int) – Number of boosted trees to fit.

gamma (float) – Minimum loss reduction required to make a further partition on a leaf node of the tree.

min_child_weight (int) – Minimum sum of instance weight(hessian) needed in a child.

subsample (float) – Subsample ratio of the training instance.

colsample_bytree (float) – Subsample ratio of columns when constructing each tree.

objective (string or callable) – Specify the learning task and the corresponding learning objective or a custom objective function to be used (see note below).

nthread (int) – Number of parallel threads used to run xgboost. (Deprecated, please use n_jobs)

scale_pos_weight (float) – Balancing of positive and negative weights.

# SVM Regressor

## Scaling data

```python
# Creating scaled set to be used in model to improve our results
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Training the model

```python
# Import SVM Regressor
from sklearn import svm

# Create a SVM Regressor
reg = svm.SVR()
```

```python
# Train the model using the training sets 
reg.fit(X_train, y_train)
```

C : float, optional (default=1.0): The penalty parameter of the error term. It controls the trade off between smooth decision boundary and classifying the training points correctly.

kernel : string, optional (default='rbf’): kernel parameters selects the type of hyperplane used to separate the data. It must be one of 'linear', 'poly', 'rbf', 'sigmoid', 'precomputed’ or a callable.

degree : int, optional (default=3): Degree of the polynomial kernel function (‘poly’). Ignored by all other kernels.

gamma : float, optional (default='auto’): It is for non linear hyperplanes. The higher the gamma value it tries to exactly fit the training data set. Current default is 'auto' which uses 1 / n_features.

coef0 : float, optional (default=0.0): Independent term in kernel function. It is only significant in 'poly' and 'sigmoid'.

shrinking : boolean, optional (default=True): Whether to use the shrinking heuristic.

## Model evaluation - Train data

```python
# Model prediction on train data
y_pred = reg.predict(X_train)
```

```python
# Model Evaluation
print('R^2:',metrics.r2_score(y_train, y_pred))
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_train, y_pred))*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1))
print('MAE:',metrics.mean_absolute_error(y_train, y_pred))
print('MSE:',metrics.mean_squared_error(y_train, y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
```
